```python
import pandas as pd
```

```python
df = pd.read_csv('Plane Crashes.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()/len(df)
```

```python
df.drop(columns='Time',axis=0)
```

```python
df = df.drop(columns=['Flight no.', 'Time'])
```

```python
df.columns
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df["Total fatalities"] = df["Total fatalities"] + 1  # Avoid log(0)
df["log_fatalities"] = np.log(df["Total fatalities"])

plt.figure(figsize=(10, 5))
sns.histplot(df["log_fatalities"], bins=30, kde=True, color="red")
plt.title("Log-Transformed Distribution of Total Fatalities")
plt.xlabel("Log(Fatalities)")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x="Fatality Group", order=["No Fatality", "1-10", "11-50", "51-100", "101-500", "500+"], palette="Reds")

# Adding labels on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight="bold")

plt.title("Categorized Fatalities per Crash")
plt.xlabel("Fatality Group")
plt.ylabel("Count")
plt.show()
```
